## Mushroom Classification Project

### Problem Description
The goal of this project is to classify mushrooms as **edible (e)** or **poisonous (p)** based on 20 descriptive features such as cap shape, cap color, stem characteristics, habitat, and season.  
This is a **binary classification problem** where the target variable is the mushroom class (`edible` or `poisonous`).

### Handle missing values: 
- **I'm applying imputation after splitting to avoid data leakage**
  
- cap-surface: impute with mode since it has skeweness

- gill-attachment: impute with mode

- gill-spacing :assign to a new category(missing) since 25063 values are null

- stem-surface: assign to a new category(missing) since 38124 values are null


```python
mode_features = ['cap-surface', 'gill-attachment']
fill_with_missing = ['gill-spacing', 'stem-surface']

# Impute mode_features using mode from training set
for col in mode_features:
    mode_value = X_train[col].mode()[0]
    X_train[col] = X_train[col].fillna(mode_value)
    X_test[col] = X_test[col].fillna(mode_value)

# Fill other features with 'missing'
for col in fill_with_missing:
    X_train[col] = X_train[col].fillna('missing')
    X_test[col] = X_test[col].fillna('missing')
```

```python
#performance before feature selection
from sklearn.base import clone
#define models
models = {
    "LogReg": LogisticRegression(max_iter=500),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier(),
    "NaiveBayes": GaussianNB()
}

results_before = []

#loop through models
for model_name, model in models.items():
    model_instance = clone(model)
    model_instance.fit(X_train, y_train)
    y_pred = model_instance.predict(X_test)
    y_prob = model_instance.predict_proba(X_test)[:, 1] 

    results_before.append({
        "Model": model_name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "F1": f1_score(y_test, y_pred),
        "ROC_AUC": roc_auc_score(y_test, y_prob)
    })

#convert to DataFrame
results_before_df = pd.DataFrame(results_before)

#print sorted results (Top 10)
print(results_before_df.sort_values(by="ROC_AUC", ascending=False).head(10))
```

```python
from sklearn.base import clone #for faster performance
#define feature selectors:
selectors= {
    "Chi2":SelectKBest(score_func=chi2, k=10),#filter
    "ANOVA":SelectKBest(score_func=f_classif, k=10),#filter
    "Mutual_Info":SelectKBest(score_func=mutual_info_classif, k=10),#filter
    "RFE_LR": RFE(LogisticRegression(max_iter=500), n_features_to_select=10),#wrapper
    "Embedded_RF": SelectFromModel(RandomForestClassifier(n_estimators=100, random_state=42), max_features=10)#embedded
}
#define classification models:
models = {
    "LogReg": LogisticRegression(max_iter=500),
    "DecisionTree": DecisionTreeClassifier(),
    "RandomForest": RandomForestClassifier(),
    "SVM": SVC(probability=True),
    "KNN": KNeighborsClassifier(),
    "NaiveBayes": GaussianNB()
}
results = []

# Loop over all combinations
for sel_name, selector in selectors.items():
    X_train_sel = selector.fit_transform(X_train, y_train)
    X_test_sel = selector.transform(X_test)

    for model_name, model in models.items():
        model_instance = clone(model)  # New instance each loop
        model_instance.fit(X_train_sel, y_train)
        y_pred = model_instance.predict(X_test_sel)
        y_prob = model_instance.predict_proba(X_test_sel)[:, 1]

        results.append({
            "Selector": sel_name,
            "Model": model_name,
            "Accuracy": accuracy_score(y_test, y_pred),
            "F1": f1_score(y_test, y_pred),
            "ROC_AUC": roc_auc_score(y_test, y_prob)
        })

results_df = pd.DataFrame(results)
print(results_df.sort_values(by="ROC_AUC", ascending=False).head(10))
```

- Using clone(model) creates a fresh, untrained instance each time, so:
  
  1. You're not carrying over state from a previously trained model.

  2. You avoid potential memory bloat or weird bugs.

  3. You ensure consistent, clean model fitting.

- Randomforest model with Embedded selector got the highest score

```python
#Get the best selector per model
best_combos = results_df.loc[results_df.groupby("Model")["ROC_AUC"].idxmax()]
print("Best combo per model:")
print(best_combos[["Model", "Selector", "Accuracy", "F1", "ROC_AUC"]])
```

- After feature reduction performance remained high for the top models but showed a very slight decline across most metrics

- insight: feature reduction slightly reduced accuracy but improved model simplicity. Feature reduction can be beneficial in reducing model complexity with minimal trade-off in performance.

```python
# evaluate only the best combos
from sklearn.metrics import classification_report, confusion_matrix

for idx, row in best_combos.iterrows():
    model_name = row["Model"]
    sel_name = row["Selector"]

    selector = selectors[sel_name]
    model = models[model_name]

    X_train_sel = selector.fit_transform(X_train, y_train)
    X_test_sel = selector.transform(X_test)

    model.fit(X_train_sel, y_train)
    y_pred = model.predict(X_test_sel)

    print(f"\nBest {model_name} with {sel_name}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4, 3))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
    plt.title(f"Confusion Matrix: {sel_name}+{model_name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Cross-validation
    from sklearn.model_selection import cross_val_score
    cv_scores = cross_val_score(model, X_train_sel, y_train, cv=5, scoring="accuracy")
    print("Cross-validation mean accuracy:", cv_scores.mean())
```

```python
def predict(model,X_train,X_test):
    y_pred_train = model.predict(X_train)
    y_pred_test= model.predict(X_test)
    return y_pred_train, y_pred_test

def evaluate(y_train, y_pred_train,y_test, y_pred_test):
    print("Training Accuracy:", accuracy_score(y_train, y_pred_train))
    print("Test Accuracy:", accuracy_score(y_test, y_pred_test))
    print("\nClassification Report:\n", classification_report(y_test, y_pred_test))
    print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_test))
```

```python
# define final models
models = {
    "Logistic Regression": (lr_final, X_test_LR_NB),
    "Naive Bayes": (nb_final, X_test_KNN_RF_SVM),
    "KNN": (knn_final, X_test_KNN_RF_SVM),
    "Decision Tree": (dt_final, X_test_DT),
    "Random Forest": (rf_final, X_test_KNN_RF_SVM), 
    "SVM": (svm_final, X_test_KNN_RF_SVM)
}

#evaluate models 
results = []
plt.figure(figsize=(10, 7))

for name, (model, X_eval) in models.items():
    # predict
    y_pred = model.predict(X_eval)
    y_proba = model.predict_proba(X_eval)[:,1] if hasattr(model, "predict_proba") else model.decision_function(X_eval)

    # metrics
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred)
    rec = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc = roc_auc_score(y_test, y_proba)

    results.append([name, acc, prec, rec, f1, roc])

    # plot ROC curve
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    plt.plot(fpr, tpr, label=f"{name} (AUC={roc:.2f})")

#ROC-AUC plot
plt.plot([0,1],[0,1],"k--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curves for All Models")
plt.legend()
plt.show()

#Results Table
results_df = pd.DataFrame(results, columns=["Model","Accuracy","Precision","Recall","F1-score","ROC-AUC"])
print("Model Comparison:")
print(results_df)

#Barplot
plt.figure(figsize=(10,6))
results_melted = results_df.melt(id_vars="Model", var_name="Metric", value_name="Score")
sns.barplot(x="Model", y="Score", hue="Metric", data=results_melted)
plt.title("Model Performance Comparison")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

- SVM (Support Vector Machine): Strong performance across the board, especially ROC-AUC (0.976), indicating excellent class separation.

Weaker Performers:

- Naive Bayes and Decision Tree perform noticeably worse, especially in recall and ROC-AUC, suggesting:
